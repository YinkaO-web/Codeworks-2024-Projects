```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#Create a random data set that has temperatures, wind speeds, precipitation
#amount, and whether or not there was a storm
weather_dictionary = {"Temperature": np.random.uniform(0, 111, 500), "Wind": np.random.uniform(7, 16, 500), "Precipitation": np.random.uniform(0, 20, 500), "Storm": np.random.randint(0, 2, 500)}
df = pd.DataFrame(weather_dictionary)
df
```

<h1>KMeans without Elbow Method</h1>

```python
#Put the data in a form in which it is (number of rows, number of columns) and
#then fit the KMeans with 3 clusters to the data. .labels_ gives us the label
#for each day
data = np.array([df['Temperature'], df['Wind'], df['Precipitation'], df['Storm']]).T
kmeans = KMeans(n_clusters = 3).fit(data)
kmeans.labels_
```

```python
plt.scatter(df['Temperature'], df['Wind'], c = kmeans.labels_, cmap = "cool")
plt.title("Temperature vs. Wind Speed for the Different Clusters")
plt.xlabel("Temperature")
plt.ylabel("Wind Speed")
plt.show()
```

<h1>KMeans with Elbow Method</h1>

```python
#loop through computing the inertias for each of the k values and adding them
#to a list to plot later
inertias = []
for i in range(1, 7):
  kmeans = KMeans(n_clusters = i).fit(data)
  inertias.append(kmeans.inertia_)
```

```python
#plot the inertias vs. the k values that are associated with them
plt.scatter(np.arange(1,7), inertias)
plt.plot(np.arange(1,7), inertias)
plt.title("K Value vs. Inertia")
plt.xlabel("K Value")
plt.ylabel("Inertia")
plt.show()
```

```python
#it appears that 2 is where the elbow appears, so group into only two this time
kmeans = KMeans(n_clusters = 2).fit(data)
kmeans.labels_
```

```python
#This gives you the locations of the different cluster centers
kmeans.cluster_centers_
```
